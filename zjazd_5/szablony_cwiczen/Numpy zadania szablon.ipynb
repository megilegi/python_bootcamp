### Czarno biały filtr

```python
im1.shape
sum_array = im1.sum(axis = 2, keepdims = True)/np.amax(sum_array)
sum_array.shape
plt.imshow(np.ones(3)*sum_array)
```

```python
sigma = 2
plt.imshow(scipy.ndimage.gaussian_filter(im1, sigma))
```
